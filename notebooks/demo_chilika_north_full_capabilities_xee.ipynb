# WetlandMapper Full Capability Demo (Chilika North, xee)

This notebook demonstrates major WetlandMapper capabilities using GEE + xee with the Chilika North shapefile.


References: Singh and Sinha (2022), Singh et al. (2022), and Sinha et al. (2017) Current Science: http://www.jstor.org/stable/24912702


```python
from pathlib import Path
import ee
import geopandas as gpd
import xarray as xr
import matplotlib.pyplot as plt
from wetlandmapper.gee import init, fetch_xee
from wetlandmapper import classify_dynamics, classify_wct_ema, classify_wct, compute_wet_frequency, aggregate_time
from wetlandmapper.terrain import compute_slope, compute_tpi, compute_local_range, mask_terrain_artifacts
from wetlandmapper.plotting import plot_wet_frequency, plot_dynamics, plot_wct, plot_index
```

```python
init()
aoi_path = Path('notebooks/chilika_north/Chilika_north.shp')
aoi_path
```

```python
mndwi_lazy = fetch_xee(aoi_path, '2000-01-01', '2023-12-31', sensor='LandsatAll', index='MNDWI', temporal_aggregation='annual', scale=30)
mndwi = mndwi_lazy.rename({'lat':'y','lon':'x'}).transpose('time','y','x').compute()
dynamics = classify_dynamics(mndwi, nYear=3, thresholdWet=25, thresholdPersis=75)
wet_freq = compute_wet_frequency(mndwi, water_threshold=0.0)
mndwi_annual = aggregate_time(mndwi, freq='annual', method='median')
mndwi, dynamics, wet_freq, mndwi_annual
```

```python
idx_lazy = fetch_xee(aoi_path, '2022-10-01', '2023-03-31', sensor='Landsat8', index=['MNDWI','NDVI','NDTI'], temporal_aggregation='seasonal', scale=30)
idx = idx_lazy.rename({'lat':'y','lon':'x'}).transpose('time','y','x').compute()
wct_ema = classify_wct_ema(idx.isel(time=0))
wct_thr = classify_wct(idx.isel(time=0))
wct_ema, wct_thr
```

```python
gdf = gpd.read_file(aoi_path).to_crs(epsg=4326)
geom = ee.Geometry(gdf.dissolve().geometry.iloc[0].__geo_interface__)
bbox = geom.bounds()
dem_img = ee.Image('COPERNICUS/DEM/GLO30').select('DEM').clip(geom)
dem_col = ee.ImageCollection([dem_img.set('system:time_start', ee.Date('2020-01-01').millis())])
dem_ds = xr.open_dataset(dem_col, engine='ee', projection=ee.Projection('EPSG:4326').atScale(30), geometry=bbox, chunks={'time':1,'lat':512,'lon':512})
dem = dem_ds['DEM'].isel(time=0).sortby('lat', ascending=False).rename({'lat':'y','lon':'x'}).compute()
slope = compute_slope(dem)
tpi = compute_tpi(dem, window=5)
local_range = compute_local_range(dem, window=5)
mndwi_masked = mask_terrain_artifacts(mndwi, dem, max_slope=5.0, max_local_range=30.0, local_range_window=5)
slope, tpi, local_range, mndwi_masked
```

```python
plot_wet_frequency(mndwi, mndwi_threshold=0.0); plt.show()
plot_dynamics(dynamics); plt.show()
plot_wct(wct_ema); plt.show()
plot_index(slope, index_name='Slope (degrees)', cmap='terrain', vmin=0); plt.show()
```
