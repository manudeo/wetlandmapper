# DEM Depression Protocol Demo

Implements the raw-vs-filled DEM depression protocol (Sinha et al., 2017, Current Science).


Reference: Sinha, R., et al. (2017). Protocols for Riverine Wetland Mapping and Classification Using Remote Sensing and GIS. Current Science, 112(7), 1544-1552. http://www.jstor.org/stable/24912702


```python
from pathlib import Path
import xarray as xr
import matplotlib.pyplot as plt
from wetlandmapper import map_dem_depressions
```

```python
raw_dem = xr.open_dataarray('data/dem_raw.tif').astype('int32')
filled_dem = xr.open_dataarray('data/dem_filled.tif').astype('int32')
raw_dem, filled_dem
```

```python
dep_raw = map_dem_depressions(raw_dem, filled_dem, apply_cleanup=False)
dep_clean = map_dem_depressions(raw_dem, filled_dem, apply_cleanup=True, cleanup_window=3, min_neighbours=2)
dep_raw, dep_clean
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 4), constrained_layout=True)
dep_raw.plot(ax=ax[0], cmap='Blues', add_colorbar=False)
ax[0].set_title('Depressions (no cleanup)')
dep_clean.plot(ax=ax[1], cmap='Blues', add_colorbar=False)
ax[1].set_title('Depressions (cleanup)')
plt.show()
```
